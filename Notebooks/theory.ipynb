### Logistic Regression

It calculates the probability of dependent variable Y, given independent variable X. 

This can be used to calculate the probability of a word having a positive or negative connotation (0, 1, or on a scale between). Or it can be used to determine the object contained in a photo (tree, flower, grass, etc.), with each object given a probability between 0 and 1.

## Basic theory: Linear vs Logistic regression

Regression is a method of modelling a target valyeu based on independent predctors. Regression technique differ based on the number of independent variables and type of relationship between indepependent and dependent variables.

Simple linear regression is a type of regression analysis where the number of independent variables is one and there is a linear relationship between the independent(x) and dependent(y) variable.

## Fit line 

A Line of best fit is a straight line that represents the best approximation of a scatter plot of data points.  It is used to study the nature of the relationship between those points. 

The equation to find the best fitting line is:

```Y` = bX + A```


where, Y` denotes the predicted value, b denotes the slope of the line, X denotes the independent variable, A is the Y intercept
